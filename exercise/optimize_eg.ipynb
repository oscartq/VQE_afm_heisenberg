```python
def energy_dummy(x):
    return (1-x[0])**2 + 100*(x[1] - x[0]**2)**2
```

```python
def callback(x):
    history_params.append(x)
    history_energy.append(energy_dummy(x))
```

```python
history_params = []
history_energy = []
init_param = np.random.rand(2)
res = optimize.minimize(energy_dummy, init_param, method="L-BFGS-B", callback=callback)
```

```python
plt.plot(history_energy)
plt.xlabel("iterations")
plt.ylabel("energy")
```
